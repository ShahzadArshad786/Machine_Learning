```python
import numpy as np 
import matplotlib.pyplot as plt
X = np.array([3, 1, 0, 4, -2])
Y = np.array([ 2, 2, 1, 3, -1 ])
a, b = np.polyfit(X, Y, 1)
```

```python
print(a)
```

```python
print(b)
```

```python
"Best Fit line is: Y = %.4f X + %.4f"%(a,b) 
```

```python
plt.rcParams['figure.figsize'] = [5, 4]
plt.scatter(X, Y)
yfit = [a * x + b  for x in X]
print("Y = " , yfit)   
plt.plot(X, yfit)
```
